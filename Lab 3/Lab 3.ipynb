# 2. Create line graphs

#### Task 1: Create a basic line graph with y values only

```python
y = np.arange(0,11,2)

fig = plt.figure()
ax = fig.subplots()
ax.plot(y)
```

#### Task 2: Create a basic line graph with x and y values

```python
x = np.arange(0,20,2)
y = np.arange(0,40,4)

fig =plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.xaxis.set(ticks=np.arange(0,18,2.5))
ax.set_ylim(-1, 38)
plt.show()
```
