# 3. Working with multiple figures and axes

#### Task 1: Create a multi-plot

```python
# initialise arrays for multi-plots
y1 = np.arange(1,10,0.5)
y2 = np.arange(1,50,5)
y3 = np.arange(100,200,10)
y4 = np.arange(1000,2000,50)
```

```python
# (a) Use the subplot method to produce a multi-plot that contains four line charts laid out in 4 rows, with following properties

fig = plt.figure(figsize = (10,15)) # plot figure

# plot chart 1 as ax1
ax1 = fig.add_subplot(411)
ax1.plot(y1,'b*')
ax1.yaxis.set(ticks = np.arange(0,10,2.5))

# plot chart 2 as ax2
ax2 = fig.add_subplot(412)
ax2.plot(y2,'s',color='#ff69b4')
ax2.yaxis.set(ticks = np.arange(0,50,20))

# plot chart 3 as ax3
ax3 = fig.add_subplot(413)
ax3.plot(y3, 'gD')
ax3.yaxis.set(ticks = np.arange(100,200,25))

# plot chart 4 as ax4
ax4 = fig.add_subplot(414)
ax4.plot(y4, linestyle='dashed', color=(1.0,0.5,0.5))
ax4.yaxis.set(ticks = np.arange(1000,2300,500))

plt.show()
```

```python
# (b) Redraw the figure such that the plots are now displayed as a 2x2 layout instead.

fig = plt.figure(figsize = (8,6)) # plot figure

# plot chart 1 as ax1
ax1 = fig.add_subplot(221)
ax1.plot(y1,'b*')
ax1.yaxis.set(ticks = np.arange(0,10,2))

# plot chart 2 as ax2
ax2 = fig.add_subplot(222)
ax2.plot(y2,'s',color='#ff69b4')
ax2.yaxis.set(ticks = np.arange(0,50,10))

# plot chart 3 as ax3
ax3 = fig.add_subplot(223)
ax3.plot(y3, 'gD')
ax3.yaxis.set(ticks = np.arange(100,200,20))

# plot chart 4 as ax4
ax4 = fig.add_subplot(224)
ax4.plot(y4, linestyle='dashed', color=(1.0,0.5,0.5))
ax4.yaxis.set(ticks = np.arange(1000,2100,200))
ax4.xaxis.set(ticks = np.arange(0,22,5))

plt.show()
```

# 4. Working with text

#### Task 1: Add text and "text-in-box"

```python
# Create the following annotated graph using the given x and y values generated using the Numpy library

x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2*np.pi*x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color="red")
ax.set_xlim(0,6)
ax.set_ylim(-1.5,1.5)

ax.text(2,1,
        'x=2, y=1')

ax.text(5,1,
        'x=5,y=1',
        bbox={'facecolor':'red'})

plt.show()
```

#### Task 2: Add annotation (arrow)

```python
# Create the following annotated graph using the given x and y values generated using the Numpy library.

x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2*np.pi*x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x,y, color = 'blue')
ax.set_ylim(-2,2)

ax.annotate('x=1.5,y=-1',
            xy=(1.5,-1),
            xytext=(0,-1.5),
            arrowprops=dict(arrowstyle='simple',
                            connectionstyle='arc3',
                            color='red'))

ax.annotate('x=2,y=1',
            xy=(2,1),
            xytext=(1,1.5),
            arrowprops=dict(arrowstyle='simple',
                            connectionstyle='arc3',
                            color='blue'))

plt.show()
```

#### Task 3: Add axis labels, subtitles and legend

```python
x = np.linspace(0, 2*np.pi, 128)
y1 = np.cos(x) # y values for c1
y2 = np.sin(x) # y values for c2

fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot(111)
ax.plot(x, y1, label = "c1", color='blue')
ax.plot(x, y2, label = "c2", color='green')

# remove plot borders
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# remove axis tick markers
ax.tick_params(length=0)

# add gridlines
ax.grid(color='w')
ax.set_facecolor('#e5e5e5')

# add legend
ax.legend(frameon=False) # legend without frame and transparent

# amend axis labels and add title and suptitle
plt.xlabel('x')
plt.ylabel('y')
plt.suptitle('MatplotLib Lab', y=1.02, fontweight='bold',fontsize=12) # use suptitle for title
plt.title('Working with text demo', fontsize=10) # use title for subtitle

plt.show()
```

# 5. Bar Charts

#### Task 1 - Analyse live births

```python
data = np.loadtxt("/Users/iylia/Library/Mobile Documents/com~apple~CloudDocs/SP/Intro to Programming for DS/Lab 3/Lab3_Datasets/live-births.csv",skiprows=1,dtype=[('year','i8'),('level_1','U50'),('value','U50')],delimiter=",")

# Extract only the rows with “Total Live-Births" in the “level_1” column
keyword = 'Total Live-Births'
column_to_search = data['level_1']
out = [i for i, v in enumerate(column_to_search) if keyword in v]
data = data[out]

# filter for years 2006 to 2016
data = data[data['year'] >= 2006]

# convert 'value' column into integer
value = data['value'].astype(int)

# plot bar chart
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

# bar chart
ax.bar(data['year'],
        value,
        color='red')

ax.yaxis.set(ticks=np.arange(0,50000,20000)) # set y-axis ticks
plt.xticks(data['year'], np.arange(2006,2017,1)) # set x-axis ticks
plt.title('Live births over the years', fontweight='bold') # set title

plt.show()
```

# 6. Pie Charts

#### Task 1 - Analyse gender composition in years 1960 vs 2016

```python
data = np.loadtxt("/Users/iylia/Library/Mobile Documents/com~apple~CloudDocs/SP/Intro to Programming for DS/Lab 3/Lab3_Datasets/singapore-residents-by-ethnic-group-and-sex-end-june-annual.csv",skiprows=1,dtype=[('year','i8'),('level_1','U50'),('value','i8')],delimiter=",")

# Extract only the rows with “Total Male/Female Residents" in the “level_1” column
data = data[np.isin(data['level_1'], ['Total Male Residents', 'Total Female Residents'])]

# plot
fig = plt.figure(figsize=(8,4.5))

# subplot 1
data_1960 = data[data['year'] == 1960]

ax1 = fig.add_subplot(121)
ax1.pie(data_1960['value'], labels=data_1960['level_1'], autopct='%1.1f%%', colors=['orange', '#2CCDBD'])
ax1.title.set_text('Gender Composition in 1960')

# subplot 2
data_2016 = data[data['year'] == 2016]

ax2 = fig.add_subplot(122)
ax2.pie(data_2016['value'], labels=data_2016['level_1'], autopct='%1.1f%%', colors=['orange', '#2CCDBD'])
ax2.title.set_text('Gender Composition in 2016')


fig.subplots_adjust(wspace=0.5)

plt.show()
```

# Task 7: Scatter Plots

#### Task 1 - Analyse relation between rainfall and temperature with scatterplot

```python
d1 = np.loadtxt("/Users/iylia/Library/Mobile Documents/com~apple~CloudDocs/SP/Intro to Programming for DS/Lab 3/Lab3_Datasets/surface-air-temperature-monthly-mean.csv",
   skiprows=1,
   dtype=[('month', 'U50'),('mean_temp','f8')],
   delimiter=',')

d2 = np.loadtxt("/Users/iylia/Library/Mobile Documents/com~apple~CloudDocs/SP/Intro to Programming for DS/Lab 3/Lab3_Datasets/rainfall-monthly-number-of-rain-days.csv",
    skiprows=1,
    dtype=[('month', 'U50'),('no_of_rainy_days','i8')],
    delimiter=',')

x = d1['mean_temp']
y = d2['no_of_rainy_days']

# plot
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x,y, color='g')

# create regresson line model for number of rainy days vs mean temp
m, b = np.polyfit(x,y,1)

# plot regression line
ax.plot(x, m*x+b, color='red')

plt.xlabel('Mean monthly temperature')
plt.ylabel('Number of rainy days')
plt.title('Monthly rainfall and temperatures since Jan 1982')

plt.show()
```

# 9. Box Plots

#### Task 1 - Analyse median rentals in different areas using box plot

```python
d = np.loadtxt('/Users/iylia/Library/Mobile Documents/com~apple~CloudDocs/SP/Intro to Programming for DS/Lab 3/Lab3_Datasets/median-rentals-and-vacancy-of-retail-space-by-locality-quarterly.csv',
    skiprows=1,
    delimiter=',',
    dtype=[('quarter', 'U50'),
            ('locality', 'U50'),
            ('ret_med_rent_lease_cm', 'f8'),
            ('ret_med_rent_cont_date', 'f8'),
            ('ret_vacancy_rate', 'f8')])

# reshape numbers data into format for box plot
data = np.array(d['ret_med_rent_lease_cm']).reshape((21,3))

# find unique elements in locality column and get index of elements (index 1 in output of this array function)
labels = np.unique(d['locality'], return_index=True)[1]

# get original order of labels based on original dataset from unique indices
label_ori_order = [d['locality'][label] for label in sorted(labels)]

# plot
fig = plt.figure(figsize=(10,9))

ax = fig.add_subplot(111)

ax.boxplot(data,
            labels=label_ori_order,
            patch_artist=True,
            medianprops={"color": "red", "linewidth": 2},
            boxprops={"facecolor": "white", "color": "blue",
                          "linewidth": 2},
            flierprops={"markerfacecolor": "green"})

ax.set_title('Median Rentals in Orchard, Outside Central Area and Central Area')

plt.show()
```
