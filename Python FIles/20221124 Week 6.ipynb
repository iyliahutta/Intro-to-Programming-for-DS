```python
import matplotlib.pyplot as plt

plt.plot(
    [2,30,40,100], # x-axis values
    [20,203,12,60]) # y-axis values
plt.ylabel('some numbers')
plt.xlabel('random numbers')
plt.axis([-20,200,0,500])
plt.show()
```

```python
import numpy as np

xvalues = [20,33,43,50,60,72,100,120,230,453]
y_1 = np.random.randint(1,50,10)
y_2 = np.random.randint(50,300,10)
plt.plot(
    xvalues,
    y_1,
    color='#32a852',
    label='Line 1'
)
plt.plot(
    xvalues,
    y_2,
    color='#a83297',
    label='Line 2'
)
legend = plt.legend()
# plt.axis([-200,800,-200,500])
plt.xlim(-200,800)
plt.ylim(-200,500)
plt.show()
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 5.0, 0.001)

plt.figure(1)
plt.subplot(131) # plt.subplot(nrow,ncol,plot number)
plt.plot(t1,'b*')

plt.subplot(132)
plt.plot(t2,'r-')

plt.subplot(133)
plt.plot(t3,'gp')
plt.show()
```
